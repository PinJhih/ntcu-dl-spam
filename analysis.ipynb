```python
import pandas as pd
import numpy as np

df = pd.read_csv("./data/enron_spam_data.csv")
df.fillna({'Message': 'missing', 'Subject': 'missing'}, inplace=True)
df
```

```python
df['m_len'] = df['Message'].str.len()
df['s_len'] = df['Subject'].str.len()
```

```python
def remove_outliers(series):
    mean = series.mean()
    std = series.std()

    lower_bound = mean - 3 * std
    upper_bound = mean + 3 * std

    series_filtered = series[(series >= lower_bound) & (series <= upper_bound)]
    removed = len(series) - len(series_filtered)
    print("Remove", removed, "in", len(series), "items.")

    return series_filtered

def print_statics(series):
    print("mean:", series.mean())
    print("std: ", series.std())
    print("quantile")
    print(series.quantile([0.25, 0.5, 0.75, 1]))
    print("===============================")
```

```python
m_len = df['m_len']
s_len = df['s_len']

print("Message String Length Statics")
print_statics(m_len)
print("Subject String Length Statics")
print_statics(s_len)
```

```python
m_len = remove_outliers(m_len)
s_len = remove_outliers(s_len)

print("Message String Length Statics after removeing outliers\n")
print_statics(m_len)
print("Subject String Length Statics after removeing outliers\n")
print_statics(s_len)
```
